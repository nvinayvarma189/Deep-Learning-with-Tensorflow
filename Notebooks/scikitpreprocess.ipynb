```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
```

```python
data = np.random.randint(0,100,(10,2))
```

```python
data
```

to put this into a nn we need to pre process this

```python
scaler_model = MinMaxScaler()
```

```python
scaler_model.fit(data) #min and max value for each column
```

```python
scaler_model.transform(data) #mapping all values of each column between zero and 1(to input into nn).....softmax
```

```python
scaler_model.fit_transform(data) #single line debba
```
